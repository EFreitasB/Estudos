```python
# exercício 1 

n = input("digite um numero: ")

try:
  if float(n) % 2 == 0:
    print(f'{n} é par')
  else:
    print(f'{n} é ímpar')
except:
  print("ERROR: O valor de entrada não é inteiro")
```

```python
# exercício 2 

h = input("digite a hora (de 0 às 23): ")

try: 
  if 0<=int(h)<=11:
    print('bom dia')
  elif 12<= int(h) <=17:
    print('boa tarde')
  elif 18<= int(h) <=23:
    print('boa noite')
  elif int(h)>23:
    print("Fora do intervalo considerado") 
except:
  print("ERROR: o valor não é inteiro")  
```

```python
# exercicio 3 

n = input("digite seu primeiro nome: ")

if len(n)<=4:
  print(f'{n} é curto')
elif 5<=len(n)<=6:
  print(f'{n} é normal')
else:
  print(f'{n} é grande')
```
