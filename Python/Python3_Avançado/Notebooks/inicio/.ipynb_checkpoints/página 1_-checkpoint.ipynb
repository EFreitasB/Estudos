## **Atividade de utilização do format e ou f'string** 

```python
nome = 'Edgar'
idade = 28 
altura = 1.88 
peso = 142.0
ano_atual = 2022 
ano_nascimento = ano_atual - 28 
imc = peso / altura ** 2 
```

```python
print(f'{nome} tem {idade} anos, {altura} de altura e pesa {peso} quilos') #cha mar as variáveis entre chaves 
print(f'O IMC de {nome} é {imc:.2f}') # aspas " "
print(f'{nome} nasceu em {ano_nascimento} ')
```

```python
print('{} tem {} anos, {:.1f} de altura e pesa {} quilos'.format(nome, idade,altura,peso)) 
```

```python
print('O IMC de {0} é {4:.2f}'.format(nome, idade,altura,peso,imc)) 
```

```python
print('{x} nasceu no ano de {y}'.format(x = nome, y = ano_nascimento))
```

```python
# encontrando palindromos

def palindromo(palavra) :
  contrario = ''

  for i in range(len(palavra)-1,-1,-1):
    contrario += palavra[i]

  if contrario == palavra:
    print('é palindromo')
  else:
    print('nao e palindromo')
```

```python
palavra = input('palavra: ') 

palindromo(palavra)
```
