```python
# benéfico quando a variavel é printada várias vezes !  
nome = 'Edgar'
print('{x} gosta de ser chamado de {x} porque {x} é o nome dele'.format(x = nome))
```

```python
# uma forma alternativa seria 

nome = 'Edgar'
print('{0} gosta de ser chamado de {0} porque {0} é o nome dele'.format(nome))
```

```python
nome = 'Magnussen da Silva' 

print(nome.upper()) 
print(nome.lower())
print(nome.title()) 
```

```python
# estamos comentando sobre manipulação de strings ! 
# cada posição (índice) guarda um valor

nome="Edgar"

nome[0:-1]
```

```python
# fatiando str
# nome[0:-1] é o mesmo que nome[:-1]

nome[:-1]
```

```python
nome[1:4]
```

```python
# do zero ao antepenultimo de 2 em 2 
nome[0:-2:2]
```

```python
# manipulando letra por letra 
# o iterador passeia pelos valores armazenados em cada índice 
nome = "Edgar"

for i in nome:
  print(i)
```
